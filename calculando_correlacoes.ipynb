# Calculator correlation
* The correlation metrify a relation inter two variables.

* The result stay inter -1 ~ 1.

> correlation = 1, the movements this variables are the same direction.

> correlation = 0, the movements this variables are don't have relations.

> correlation = -1, the movements this variables are the oposite direction.

* For one example, imagine to we want to calculation a relation inter ITUB4 and BBAS3, likely the correlation go be very agressive, one time this is tow banks.

### Here, you import the libs who the you go to the use in this application. 

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web

import matplotlib.pyplot as plt
import investpy as inv

from scipy import stats

import yfinance as yf
```

## Choosing the stocks

* How we are picking stocks the yahoo finance, it's codes are more .SA.  

* In Case you want choice others stocks, enough change his names in the code, or puth more others.

```python
papeis = ['ITUB4.SA', 'PETR4.SA', 'VVAR3.SA', 'BBDC4.SA', 'VALE3.SA', 'CIEL3.SA', 
          'CSAN3.SA', 'ABEV3.SA', 'AZUL4.SA', 'BTOW3.SA', 'B3SA3.SA', 'BBAS3.SA', '^BVSP']
PAPEIS_ANALISADOS = pd.DataFrame()
for papel in papeis:
    PAPEIS_ANALISADOS[papel] = web.DataReader(papel, data_source= 'yahoo',start= '01/01/2015')['Close']
```

```python
correlacoes = PAPEIS_ANALISADOS.corr()
```

## Here are the configs the correlations graph. And after that the correlation graph. 

```python
plt.figure(figsize=(10,7))
plt.imshow(correlacoes, cmap="RdYlGn", interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(correlacoes)), correlacoes.columns, rotation = 'vertical')
plt.yticks(range(len(correlacoes)), correlacoes.columns)
plt.suptitle("MAPA DE CORRELAÇÕES - RETORNO", fontsize = 15, fontweight = 'bold')
correlacoes_grafico = plt.gcf()
```
